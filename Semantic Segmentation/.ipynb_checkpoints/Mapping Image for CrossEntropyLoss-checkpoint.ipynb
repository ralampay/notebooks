```python
import torch
import numpy as np
import os
import sys
import cv2
from matplotlib import pyplot as plt
```

```python
n_classes = 32
img_width = 200
img_height = 200
dim = (img_width, img_height)
test_image_file = "/home/ralampay/Pictures/CamVid/train_labels/0001TP_009210_L.png"
```

```python
test_image = cv2.resize(cv2.imread(test_image_file, 0), dim)

print(test_image.shape)
plt.imshow(test_image)
plt.show()
```

```python
def map_pixel(image, n_classes):
    bin_width = np.round(256 / n_classes)
    print("bin_width: {}".format(bin_width))
    labels = []
    
    for i in range(n_classes):
        if i == 0:
            min_val = 0
        else:
            min_val = bin_width * i
            
        max_val = min_val + bin_width
        
        labels.append([i, min_val, max_val])
        
    h = image.shape[0]
    w = image.shape[1]
    
    for y in range(0, h):
        for x in range(0, w):
            val = image[y, x]
            
            for label in labels:
                if val >= label[1] and val < label[2]:
                    image[y, x] = label[0]
        
    return image
result = map_pixel(test_image, n_classes)
```
